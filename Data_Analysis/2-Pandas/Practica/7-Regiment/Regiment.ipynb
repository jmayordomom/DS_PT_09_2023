### Paso 2. Cree el DataFrame con los siguientes valores:

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

### Paso 3. Asígnalo a una variable llamada regimiento.
##### No olvides nombrar cada columna

```python
regimiento = pd.DataFrame(raw_data)
regimiento
```

### Paso 4. ¿Cuál es la puntuación media preTestScore del regimiento Nighthawks?  

```python
media = regimiento[regimiento["regiment"] == "Nighthawks"]["preTestScore"].mean()
print(f"La media de Nighthawks para el preTestScore es {media}")
```

### Paso 5. Presentar estadísticas generales por empresa

```python
regimiento.groupby("company").describe().round(2)
```

### Paso 6. ¿Cuál es la media de la puntuación preTest de cada empresa?

```python
regimiento.groupby("company").mean("preTestScore").round(2)
```

### Paso 7. Presentar las puntuaciones medias preTest agrupadas por regimiento y compañía.

```python
regimiento.groupby(["regiment","company"]).mean("preTestScore")["preTestScore"].round(2)
```

### Paso 8. Presentar las puntuaciones medias preTestScores agrupadas por regimiento y compañía sin indexación heirárquica.

```python
regimiento.groupby(["regiment","company"]).mean("preTestScore")["preTestScore"].unstack()
```

### Paso 9. Agrupar todo el marco de datos por regimiento y compañía

### Paso 11. Iterar sobre un grupo e imprimir el nombre y todos los datos del regimiento. Haz una función llamada agrupa(variable)

```python
#crear una función que se le pasa un argumento, y por ese argumento se hace group by y se muestra la info asociada
def agrupacion(df, variable):
    res = df.groupby(variable).mean(variable)
    print("Media del regimiento:",res, sep="\n")
    print()
    res = df.groupby(variable).median(variable)
    print("Mediana del regimiento:",res, sep="\n")
    print()

agrupacion(regimiento, "regiment")
```
