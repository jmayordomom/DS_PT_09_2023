# Occupation

### Paso 1. Importar librerias

```python
import pandas as pd
```

### Paso 2. Importar el conjunto de datos de esta [dirección](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
url ="https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

### Paso 3. Asígnalo a una variable llamada usuarios.

```python
usuarios = pd.read_csv(url, sep="|")
usuarios = usuarios.drop("user_id", axis =1)
usuarios
```

### Paso 4. Descubre cuál es la edad media por ocupación

```python
usuarios.groupby("occupation").mean("age")["age"]
```

### Paso 5. Descubra la proporción de Hombres por ocupación y ordénela de mayor a menor

#### Forma 1

```python
total = usuarios[["gender","occupation"]].groupby("occupation").count().sort_values(by="occupation")
total
```

```python
totalHombres = usuarios[usuarios["gender"] == "M"][["gender","occupation"]].groupby("occupation").count().sort_values(by="occupation")
totalHombres
```

```python
for i in range(len(total)):
    ocupacion = total.iloc[i].name
    print("Ocupación",ocupacion)
    porcentaje = totalHombres.iloc[i].values[0]/total.iloc[i].values[0]
    print(f"La proporción de hombres en {ocupacion} es del {round(porcentaje*100,2)}%")
    print()
```

### Paso 6. Para cada ocupación, calcule las edades mínima y máxima

### Paso 7. Para cada combinación de ocupación y sexo, calcule la edad media

```python
usuarios.groupby(by=["occupation", "gender"]).mean(numeric_only=True).round(2)
```

### Paso 8.  Para cada ocupación, presente el porcentaje de mujeres y hombres

```python
for i in range(len(total)):
    ocupacion = total.iloc[i].name
    print("Ocupación",ocupacion)
    porcentaje = totalHombres.iloc[i].values[0]/total.iloc[i].values[0]
    print(f"La proporción de hombres en {ocupacion} es del {round(porcentaje*100,2)}%")
    porcMujeres = 100-round(porcentaje*100,2)
    print(f"La proporción de mujeres en {ocupacion} es del {round(porcMujeres,2)}%")
    print()
```
